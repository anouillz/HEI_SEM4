```python
import pandas as pd
```

```python
# Load the data
data = pd.read_csv('data_section2/supermarket_sales-Sheet1.csv')
```

```python
# only Yangon city
city_condition = (data['City'] == 'Yangon')
data_filtered = data.loc[city_condition]
```

```python
data_filtered
```

```python
# transactions conditions
low_trans_condition = (data_filtered['Rating'] > 4.0) 
high_trans_condition = (data_filtered['Rating'] < 9.9)
data_filtered = data_filtered.loc[low_trans_condition]
data_filtered = data_filtered.loc[high_trans_condition]
```

```python
# total price in new column ( unit price * quantity + 5% tax)
data_filtered['Total_Price'] = data_filtered['Unit price'] * data_filtered['Quantity'] + data_filtered['Tax 5%']
```

```python
# show columns total and total price
data_filtered[['Total', 'Total_Price']]
```

```python
# apply 10% discount on total price but only for members
data_filtered['Final Price'] = data_filtered.apply(lambda x: x['Total_Price'] * 0.9 if x['Customer type'] == 'Member' else x['Total_Price'], axis=1)
```
