```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('medals_long.csv')
df.head()
```

```python
# total medals for each nation
total_medals = df.groupby('nation')['count'].sum()
total_medals
```

```python
sns.set_theme()
plt.figure(figsize=(10, 6))

medal_colors = ['brown', 'gold', 'gray']

crosstb = pd.crosstab(index=df.nation, columns=df.medal, values=df['count'], aggfunc='sum')
crosstb.plot(kind='bar', stacked=True, color = medal_colors)


plt.xlabel('Nation')
plt.ylabel('Count')
plt.title('Total Medals by Nation')

plt.show()
```

```python
sns.set_theme()
plt.figure(figsize=(10, 6))

barplt = crosstb.plot(kind='bar', stacked=False, color = medal_colors)

barplt.set_ylabel('Count')
barplt.set_xlabel('Nation')
barplt.set_title('Total Medals by Nation')

plt.show()
```
