```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.pyplot import xlabel, ylabel
```

```python
df = pd.read_csv('iris.csv')
df.head()
```

```python
# scatter plot matrix of df 
sns.pairplot(df, hue='variety')

plt.legend = True

plt.show()
```

```python
# density estimations of df
sns.set_theme()
sns.kdeplot(df['sepal.length'])
```

```python
# parallel coordinates plot of df
pd.plotting.parallel_coordinates(df, 'variety')
```

```python
# heatmap of df without variety
sns.heatmap(df['variety'])
```
