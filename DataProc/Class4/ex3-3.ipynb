```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('iris.csv')
df.head()
```

```python
# histogram of all sepal lengths
df['sepal.length'].plot.hist()

plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.title('Histogram of Sepal Length')
```

```python
# 4 sublots, 1 per feature, overlaid on one plot
s_length_hist = df['sepal.length'].plot.hist(alpha=0.5, label='Sepal Length')
p_length_hist = df['petal.length'].plot.hist(alpha=0.5, label='Petal Length')
p_width_hist = df['petal.width'].plot.hist(alpha=0.5, label='Petal Width')
s_width_hist = df['sepal.width'].plot.hist(alpha=0.5, label='Sepal Width')
```

```python
p_length = df.groupby('variety')['petal.length'].plot.hist(alpha= 0.5, xlabel = 'Petal Length', ylabel = 'Frequency', title = 'Histogram of Petal Length', legend = True)
```
